```python
from urllib.parse import urlparse
import requests
import pandas as pd
from tqdm import tqdm
import os
import glob
import json
```

```python
token = "INSERT_TOKEN"
headers = {"Authorization": f"token {token}"}
```

```python
all_fix_prs_with_issues = pd.read_parquet('output_files/fix_PRs_with_issues.parquet')
```

```python
def extract_owner_repo(url):
    path = url.strip().replace("https://github.com/", "").replace("https://api.github.com/repos/", "")
    parts = path.strip("/").split("/")
    return parts[0], parts[1]
```

```python
# --- Fun√ß√£o para buscar arquivos modificados (com status e patch)
def get_pr_files_detailed(owner, repo, pr_number, headers):
    """Retorna lista de dicion√°rios: filename, status e patch"""
    files_info = []
    page = 1
    while True:
        url = f"https://api.github.com/repos/{owner}/{repo}/pulls/{pr_number}/files?page={page}&per_page=100"
        r = requests.get(url, headers=headers)
        if r.status_code != 200:
            print(f"‚ö†Ô∏è Erro {r.status_code} no PR {pr_number} ({owner}/{repo})")
            break
        data = r.json()
        if not data:
            break

        for f in data:
            files_info.append({
                "filename": f.get("filename"),
                "status": f.get("status"),
                "additions": f.get("additions"),
                "deletions": f.get("deletions"),
                "changes": f.get("changes"),
                "patch": f.get("patch")
            })
        page += 1
    return files_info
```

```python
def safe_json_load(x):
    """
    Fun√ß√£o auxiliar para carregar JSON de forma segura de uma c√©lula do parquet, 
    tratando c√©lulas vazias/NaN ou JSON mal formatado.
    """
    if pd.isna(x):
        return [] # Retorna lista vazia se a c√©lula era NaN
    
    try:
        # Tenta carregar como JSON (formato padr√£o e correto)
        return json.loads(x)
    except (json.JSONDecodeError, TypeError):
        # Se falhar, tenta como um 'literal' do Python (ex: "['a', 'b']")
        # Isso pode acontecer se o parquet foi salvo de forma estranha
        try:
            import ast
            result = ast.literal_eval(x)
            # Garante que o resultado √© de fato uma lista
            return result if isinstance(result, list) else []
        except (ValueError, SyntaxError):
            # Se ambos falharem, √© um dado inv√°lido
            print(f"Aviso: Falha ao decodificar JSON/literal, tratando como vazio: {str(x)[:50]}...")
            return [] # Retorna lista vazia se for inv√°lido
```

```python
# --- Configura√ß√£o ---
save_interval = 50
final_column_name = "modified_files" # Nome da coluna de resultados
partial_file_prefix = "output_files/partial/prs_with_files_partial_"
final_output_file = "output_files/prs_with_files_FINAL.parquet" # MUDADO PARA .parquet
results = []
start_index = 0
df_final_encontrado = False # Flag para pular o loop

# ===================================================
# 0. VERIFICAR SE O ARQUIVO FINAL J√Å EXISTE (NOVO BLOCO)
# (Assume que 'final_output_file' = "prs_with_files_FINAL.parquet")
# ===================================================
if os.path.exists(final_output_file):
    print(f"üéâ Encontrado arquivo final: {final_output_file}")
    print("Carregando resultados finais e pulando o loop...")
    
    try:
        # Carregamos o DF final
        df_completo = pd.read_parquet(final_output_file)
        
        # Deserializamos a coluna (assume 'final_column_name' = "modified_files")
        df_completo[final_column_name] = df_completo[final_column_name].apply(safe_json_load)
        
        # Sincronizamos 'results' e 'start_index' para pular o loop
        results = df_completo[final_column_name].tolist()
        start_index = len(df_completo) 
        
        # IMPORTANTE: Sobrescrevemos o 'all_fix_prs_with_issues' com a vers√£o completa
        all_fix_prs_with_issues = df_completo
        df_final_encontrado = True # Seta a flag
        
        print(f"‚úÖ Resultados finais carregados. {len(results)} linhas.")
    
    except Exception as e:
        print(f"‚ö†Ô∏è Erro ao carregar o arquivo final {final_output_file}: {e}")
        print("Ignorando e continuando com a l√≥gica de retomada parcial...")
```

```python


# ===================================================
# 1. PROCURAR ARQUIVO PARCIAL (se o final n√£o foi carregado)
# ===================================================
if not df_final_encontrado: # S√≥ executa se o carregamento final falhou ou n√£o existia
    partial_files = glob.glob(f"{partial_file_prefix}*.parquet") 

    if partial_files:
        latest_partial_file = max(partial_files, key=os.path.getmtime)
        print(f"Encontrado salvamento parcial: {latest_partial_file}")
        
        try:
            df_partial = pd.read_parquet(latest_partial_file) 
            
            # Deserializar o JSON
            results = df_partial[final_column_name].apply(safe_json_load).tolist()
            start_index = len(results)                       
            
            print(f"Carregados {start_index} resultados anteriores (de parquet).")
            print(f"Retomando o processo a partir do √≠ndice {start_index}...")

        except Exception as e:
            print(f"Erro ao carregar ou processar {latest_partial_file}: {e}. Come√ßando do zero.")
            results = []
            start_index = 0
    else:
        print("Nenhum salvamento parcial encontrado. Come√ßando do zero.")

# --- Loop principal Modificado ---

# 4. N√≥s iteramos apenas sobre o que FALTA
df_remaining = all_fix_prs_with_issues.iloc[start_index:]

print(f"Processando {len(df_remaining)} PRs restantes (de {start_index} at√© {len(all_fix_prs_with_issues)})...")

# 5. Usamos 'start=start_index' no enumerate
for i, (idx, row) in enumerate(tqdm(df_remaining.iterrows(), total=len(df_remaining)), start=start_index):
    owner, repo = extract_owner_repo(row["repo_url"])
    pr_number = row["number"]

    try:
        files = get_pr_files_detailed(owner, repo, pr_number, headers)
    except Exception as e:
        print(f"Erro no PR {pr_number} ({owner}/{repo}): {e}")
        files = [] # O valor de erro deve ser uma lista vazia

    # 1. Apenas colete na lista:
    results.append(files) # Adiciona o *novo* resultado (ainda como lista Python)
    
    # üíæ Salvamento parcial
    if (i + 1) % save_interval == 0:
        print(f"\nIniciando salvamento parcial (processados {i+1})...")
        
        # 2. CRIE UM DATAFRAME PARCIAL
        df_partial = all_fix_prs_with_issues.iloc[:i+1].copy()  
        
        # 3. ATRIBUA A LISTA DE RESULTADOS
        df_partial[final_column_name] = results
        
        # 4. SERIALIZE A COLUNA PARA JSON (NOVO PASSO!)
        # O parquet n√£o pode guardar listas, ent√£o transformamos a lista em uma string JSON.
        df_partial[final_column_name] = df_partial[final_column_name].apply(json.dumps)
        
        # 5. SALVE O DF PARCIAL (em parquet)
        partial_path = f"{partial_file_prefix}{i+1}.parquet" # MUDADO PARA .parquet
        df_partial.to_parquet(partial_path, index=False) # MUDADO PARA .to_parquet
        print(f"‚úÖ Salvamento parcial: {partial_path}")


# --- Salvamento final
print("\nLoop conclu√≠do. Preparando salvamento final...")

#display(results[0])

# 5. ATRIBUI√á√ÉO FINAL
all_fix_prs_with_issues[final_column_name] = results
# 6. SERIALIZA√á√ÉO FINAL (NOVO PASSO!)
# Tamb√©m precisamos serializar a coluna final antes de salvar em parquet
all_fix_prs_with_issues[final_column_name] = all_fix_prs_with_issues[final_column_name].apply(json.dumps)

# 7. SALVAMENTO FINAL
all_fix_prs_with_issues.to_parquet(final_output_file, index=False) # MUDADO PARA .to_parquet
print(f"‚úÖ Salvamento final completo: {final_output_file}")
```

```python
# ===================================================
# üßπ BLOCO DE LIMPEZA (COLE A PARTIR DAQUI)
# ===================================================
print(f"\nLimpando arquivos parciais com prefixo: '{partial_file_prefix}'...")
try:
    # partial_file_prefix = "prs_with_files_partial_"
    # Certifique-se de que a extens√£o √© a correta (.parquet ou .pkl)
    pattern_to_clean = f"{partial_file_prefix}*.parquet" 
    
    # Se voc√™ ainda estiver usando pickle para este script:
    # pattern_to_clean = f"{partial_file_prefix}*.pkl" 
    
    files_to_delete = glob.glob(pattern_to_clean)
    
    if not files_to_delete:
        print("Nenhum arquivo parcial para limpar.")
    else:
        for f in files_to_delete:
            os.remove(f)
        print(f"‚úÖ {len(files_to_delete)} arquivos parciais removidos.")

except Exception as e:
    print(f"‚ö†Ô∏è Erro durante a limpeza dos arquivos parciais: {e}")
# ===================================================
# FIM DO BLOCO DE LIMPEZA
# ===================================================

print("\nProcesso conclu√≠do.")
```

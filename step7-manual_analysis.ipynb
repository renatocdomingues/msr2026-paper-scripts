```python
import pandas as pd
```

```python
file1 = pd.read_csv(r'output_files/comments/comments_extracted_from_review_type.csv')
file2 = pd.read_csv(r'output_files/comments/comments_extracted_from_review_comment.csv')
file3 = pd.read_csv(r'output_files/comments/comments_extracted_from_pr_comment.csv')
```

```python
columns = ['id','number','agent','html_url']
df_concat1 = pd.concat([file1[columns], file2[columns]])
df_concat = pd.concat([df_concat1[columns], file3[columns]])
prs = df_concat.drop_duplicates()
sample = prs.sample(n=50, random_state=50)
sample
```

```python
sample.to_csv(r'output_files/comments/50_sample.csv', index = False)
```
